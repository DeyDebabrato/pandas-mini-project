```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"E:\AI_CLASS\MiniProject_Pandas\data\raw\us-shein-appliances-3987.csv")
df.head()
```

```python
print("\nShape:", df.shape)
print("\nColumns:", df.columns)
print("\nData Types:\n", df.dtypes)

df.info()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].astype(str).str.replace("$", "", regex=False)
        df[col] = df[col].astype(str).str.replace(",", "", regex=False)
```

```python
for col in df.columns:
    try:
        df[col] = pd.to_numeric(df[col])
    except:
        pass

# Check missing values
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.drop_duplicates(inplace=True)

print("\nCleaning Completed")
```

```python
important_cols = [
    col for col in df.columns
    if "price" in col or "rating" in col or "review" in col
]

print("\nImportant Columns Preview:")
print(df[important_cols].head())
```

```python
if "original_price" in df.columns and "sale_price" in df.columns:
    df["discount"] = df["original_price"] - df["sale_price"]
    print("\nDiscount column created")
```

```python
if "rating" in df.columns:
    high_rated = df[df["rating"] >= 4]
    print("\nHigh Rated Products:")
    print(high_rated.head())
```

```python
if "rating" in df.columns:
    print("\nTop Rated Products:")
    print(df.sort_values(by="rating", ascending=False).head())
```

```python
if "category" in df.columns and "rating" in df.columns:
    print("\nAverage Rating by Category:")
    print(df.groupby("category")["rating"].mean())
```

```python
if "category" in df.columns and "review_count" in df.columns:
    print("\nTotal Reviews by Category:")
    print(df.groupby("category")["review_count"].sum())
```

```python
if "category" in df.columns:
    numeric_cols = df.select_dtypes(include=np.number).columns
    print("\nNumeric Analysis by Category:")
    print(df.groupby("category")[numeric_cols].mean())
```

```python
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
df.to_csv("../data/processed/cleaned_shein_appliances.csv", index=False)
```
